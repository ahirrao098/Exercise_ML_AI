```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

```python
sns.distplot(df['per capita income (US$)'])
```

```python
df=df.rename({'per capita income (US$)':'income'},axis=1)
```

```python
df
```

```python
from sklearn import linear_model
```

```python
plt.xlabel('income')
plt.ylabel('year')
plt.scatter(df.income,df.year,color='red',marker='+')
```

```python
new_df=df.drop('income',axis='columns')
new_df
```

```python
income=df.income
income
```

```python
# Create linear regression object
reg=linear_model.LinearRegression()
reg.fit(new_df,income)
```

```python
#predict the per capita income fo canadian citizens in year 2020
```

```python
reg.predict([[2020]])
```

```python
df.corr()
```

different method of SLR

```python
import statsmodels.formula.api as smf
model=smf.ols('income~year',data=df).fit()
```

```python
model.params
```

```python
print(model.tvalues, '\n', model.pvalues)
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

```python
newdata =pd.Series([2020,2021])
```

```python
data_pred=pd.DataFrame(newdata,columns=['year'])
```

```python
model.predict(data_pred)
```
