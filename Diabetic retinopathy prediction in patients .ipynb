## Using sklearn OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()#class object
```

```python
df.prognosis=le.fit_transform(df.prognosis)
df
```

```python
retinopathy=df[df.prognosis==1]
retinopathy.shape
```

```python
no_retinopathy=df[df.prognosis==0]
no_retinopathy.shape
```

```python
df.groupby('prognosis').mean()
```

### Dummy Variable

```python
prognosis_dummies=pd.get_dummies(df.prognosis)
prognosis_dummies
```

```python
df_with_dummies = pd.concat([df,prognosis_dummies],axis='columns')
df_with_dummies
```

```python
df_with_dummies.drop('prognosis',axis='columns',inplace=True)
df_with_dummies.head()
```
